### 암 2,3,4기에 있어서, 수술 후 항암제를 사용한 그룹과 사용하지 않은 그룹의 생존 기간 차이가 있는가?

```python
ptnm_4_0 = Postop_Chemotherapy_0[(Postop_Chemotherapy_0['pTNM'] == 4) | 
                      (Postop_Chemotherapy_0['pTNM'] == 41) | 
                      (Postop_Chemotherapy_0['pTNM'] == 42) | 
                      (Postop_Chemotherapy_0['pTNM'] == 43)]

ptnm_4_1 = Postop_Chemotherapy_1[(Postop_Chemotherapy_1['pTNM'] == 4) | 
                      (Postop_Chemotherapy_1['pTNM'] == 41) | 
                      (Postop_Chemotherapy_1['pTNM'] == 42) | 
                      (Postop_Chemotherapy_1['pTNM'] == 43)]
stats.ttest_ind(ptnm_4_0['OS'], ptnm_4_1['OS'])
```

```python
ptnm_3_0 = Postop_Chemotherapy_0[(Postop_Chemotherapy_0['pTNM'] == 3) | 
                      (Postop_Chemotherapy_0['pTNM'] == 31) | 
                      (Postop_Chemotherapy_0['pTNM'] == 32) | 
                      (Postop_Chemotherapy_0['pTNM'] == 33)]

ptnm_3_1 = Postop_Chemotherapy_1[(Postop_Chemotherapy_1['pTNM'] == 3) | 
                      (Postop_Chemotherapy_1['pTNM'] == 31) | 
                      (Postop_Chemotherapy_1['pTNM'] == 32) | 
                      (Postop_Chemotherapy_1['pTNM'] == 33)]
stats.ttest_ind(ptnm_3_0['OS'], ptnm_3_1['OS'])
```

```python
ptnm_2_0 = Postop_Chemotherapy_0[(Postop_Chemotherapy_0['pTNM'] == 2) | 
                      (Postop_Chemotherapy_0['pTNM'] == 21) | 
                      (Postop_Chemotherapy_0['pTNM'] == 22) | 
                      (Postop_Chemotherapy_0['pTNM'] == 23)]

ptnm_2_1 = Postop_Chemotherapy_1[(Postop_Chemotherapy_1['pTNM'] == 2) | 
                      (Postop_Chemotherapy_1['pTNM'] == 21) | 
                      (Postop_Chemotherapy_1['pTNM'] == 22) | 
                      (Postop_Chemotherapy_1['pTNM'] == 23)]
stats.ttest_ind(ptnm_2_0['OS'], ptnm_2_1['OS'])
```

### 암 기수에 있어서, 수술 후 항암제를 어떤 것으로 사용하였는지 따라, 생존 기간 차이가 있는가?

```python
df['pTNM'].value_counts()
```

### 2기

```python
df2 = df[(df['pTNM'] == 2) |
         (df['pTNM'] == 21) |
   (df['pTNM'] == 22) |
   (df['pTNM'] == 23) ]
```

```python
df2['Postop_Chemo_Regimen'].value_counts()
```

```python
# the "C" indicates categorical data
model = ols('OS ~ C(Postop_Chemo_Regimen)', df2).fit()

print(anova_lm(model))
```

### 3기

```python
df3 = df[(df['pTNM'] == 3) |
    (df['pTNM'] == 31) |
   (df['pTNM'] == 32) |
   (df['pTNM'] == 33) ]
# the "C" indicates categorical data
model = ols('OS ~ C(Postop_Chemo_Regimen)', df3).fit()

print(anova_lm(model))   # 0.00000000232
```

### 4기

```python
df4 = df[(df['pTNM'] == 4) |
    (df['pTNM'] == 41) |
   (df['pTNM'] == 42) |
   (df['pTNM'] == 43) ]

# the "C" indicates categorical data
model = ols('OS ~ C(Postop_Chemo_Regimen)', df4).fit()

print(anova_lm(model))   # 0.00000000232
```

결론적으로, 수술 후 항암제의 종류에 따라, 생존기간의 차이가 있었다.

## 직장암과 결장암

```python
rectal = df[df['Location_of_rectal_cancer'].notnull()]
```

```python
stats.ttest_ind(colon['OS'], rectal['OS']) #의미 없음
```

```python
rectal['Postop_Chemo_Regimen'].value_counts()  #5-FU/L
```

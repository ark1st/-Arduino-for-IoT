```python
#어떤 성별이 팁을 많이 주는가
df1 = tips.groupby(['sex']).mean()
df1
```

```python
df2 = tips.groupby(['day']).mean()
df2
```

```python
df3 = tips.groupby(['day']).count()
df3
```

```python
df4 = tips.groupby(['day']).describe()

df4.T
```

### 다변수분석

```python
#성별 별/ 흡연자가 얼마나 팁/금액을 지불했는가?

df5 = tips.groupby(['sex', 'smoker']).mean()
df5
```

```python
# 요일별, 시간대에 따라서.

df6 = tips.groupby(['day','time']).mean()
df6
```
